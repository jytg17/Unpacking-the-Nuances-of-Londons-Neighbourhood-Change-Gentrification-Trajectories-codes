```python
# importing pandas library

import pandas as pd
```

```python
# converting float64 columns which contain integers to int64 for consistency

original_df['pop_2001'] = original_df['pop_2001'].astype('int64')
original_df['All_cat_qual_2001'] = original_df['All_cat_qual_2001'].astype('int64')
original_df['Level_4_qualifications_and_above_2001'] = original_df['Level_4_qualifications_and_above_2001'].astype('int64')
original_df['All_categories_NS_SeC_2001'] = original_df['All_categories_NS_SeC_2001'].astype('int64')
original_df['X1_Higher_managerial_administrative_and_professional_occupations_2001'] = original_df['X1_Higher_managerial_administrative_and_professional_occupations_2001'].astype('int64')
```

```python
# re-checking the column types

original_df.info()
```

```python
original_df.head()
```

```python
# standardise private rental, qualifications and NS-sec values into percentage rates / proportions

# calculate the incidence of residents with Lvl 4 qual & above as a proportion of the overall number of residents across all qual categories
original_df['Level_4_qualifications_and_above_2001_pct'] = (original_df['Level_4_qualifications_and_above_2001'] / original_df['All_cat_qual_2001']) * 100
original_df['Level_4_qualifications_and_above_2011_pct'] = (original_df['Level_4_qualifications_and_above_2011'] / original_df['All_cat_qual_2011']) * 100

# calculate the incidence of residents in the top NS-sec tier (Higer Managerial, Administrative and Professional Occupations) as a proportion of the overall number of residents across all NS-sec categories
original_df['X1_Higher_managerial_administrative_and_professional_occupations_2001_pct'] = (original_df['X1_Higher_managerial_administrative_and_professional_occupations_2001'] / original_df['All_categories_NS_SeC_2001']) * 100
original_df['X1_Higher_managerial_administrative_and_professional_occupations_2011_pct'] = (original_df['X1_Higher_managerial_administrative_and_professional_occupations_2011'] / original_df['All_categories_NS_SeC_2011']) * 100
```

## Scaling the data as prep for PCA to derive Composite Index scores for all LSOAs

```python
# import necessary libraries for PCA

from sklearn.preprocessing import RobustScaler
from sklearn import decomposition  
from sklearn.preprocessing import scale  
from sklearn import preprocessing 
from sklearn import linear_model
from sklearn import model_selection
import numpy as np

df_scaling = extract_df.set_index('LSOA_Code')
```

```python
# check that index has been changed to LSOA code

df_scaling.head()
```

```python
# split 2001 data from 2011 data

df_scaling_2001 = df_scaling[["Median_House_Prices_2001", "Level_4_qualifications_and_above_2001_pct", "X1_Higher_managerial_administrative_and_professional_occupations_2001_pct", "Median_income_2001"]]
df_scaling_2011 = df_scaling[["Median_House_Prices_2011", "Level_4_qualifications_and_above_2011_pct", "X1_Higher_managerial_administrative_and_professional_occupations_2011_pct", "Median_income_2011"]]
```

```python
# stacking the 2001 data vertically with 2011 data

values_2001 = df_scaling_2001.values
values_2011 = df_scaling_2011.values
stacked_df = np.concatenate((values_2001, values_2011), axis=0)
```

```python
# checking for NaN values before scaling and transforming

np.isnan(stacked_df).sum()
```

```python
# scaling and transforming the stacked data using RobustScaler

Robust_transformer = RobustScaler().fit(stacked_df)
Robust_transformer_df = Robust_transformer.transform(stacked_df)
Robust_transformer_df
```

```python
# extracting scaled data for 2001

scaled_df_2001 = pd.DataFrame(Robust_transformer_df[0:len(df_scaling_2001),], index=df_scaling_2001.index, columns=list(df_scaling_2001.columns))
scaled_df_2001
```

```python
# extracting scaled data for 2011

scaled_df_2011 = pd.DataFrame(Robust_transformer_df[len(df_scaling_2011):,], index=df_scaling_2011.index, columns=list(df_scaling_2011.columns))
scaled_df_2011
```

```python
# visualising scaled data using boxplot (2001)

scaled_df_2001 = scaled_df_2001.rename(index=str, columns={"Median_House_Prices_2001": "hseP_01", 
                                                   "Level_4_qualifications_and_above_2001_pct": "LV4_01",
                                                   "X1_Higher_managerial_administrative_and_professional_occupations_2001_pct": "SEC1_01",
                                                   "Median_income_2001": "inc_01"})
scaled_df_2001.boxplot()

# to output high-resolution image - uncomment the next line
# plt.savefig('scaled_df_2001.png', dpi = 100)
```

```python
# visualising scaled data using boxplot (2011)

scaled_df_2011 = scaled_df_2011.rename(index=str, columns={"Median_House_Prices_2011": "hseP_11", 
                                                   "Level_4_qualifications_and_above_2011_pct": "LV4_11",
                                                   "X1_Higher_managerial_administrative_and_professional_occupations_2011_pct": "SEC1_11",
                                                   "Median_income_2011": "inc_11"})

scaled_df_2011.boxplot()

# to output high-resolution image - uncomment the next line
# plt.savefig('scaled_df_2011.png', dpi = 100)
```

```python
# carry out PCA to derive only the 1st principal component

pca = decomposition.PCA(n_components=1)             
pca.fit(Robust_transformer_df)                      
PCA_scores = pd.DataFrame(pca.transform(Robust_transformer_df))  

print("Explained variance of the SES score is: {0:6.5f}".format(pca.explained_variance_ratio_[0]))
```

```python
# view Composite Scores

PCA_scores.head()
```

```python
# what if one wanted to find all principal components?

pca_all = decomposition.PCA()             
pca_all.fit(Robust_transformer_df)                      
print(pca_all.explained_variance_ratio_)
```

```python
# visualise explained variance for each PC

plt.bar(['PC 1', 'PC 2', 'PC 3', 'PC 4'],pca_all.explained_variance_ratio_)

x = [1,2,3,4]

for i in range(len(pca_all.explained_variance_ratio_)):
    plt.text(x = x[i]-1.2, y = pca_all.explained_variance_ratio_[i] + 0.01, s = pca_all.explained_variance_ratio_[i].round(3), size = 12)
    
plt.xlabel('Principal Components (PC)', fontsize=12)
plt.ylabel('Explained Variance', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
# visualise the constituents of each PC using heatmap

features = ["Median House Price", "Lvl 4 Qual", "NS-SEC 1", "Median Income"]

plt.matshow(pca_all.components_,cmap='viridis')
plt.yticks([0,1,2,3],['PC 1','PC 2','PC 3', 'PC 4'],fontsize=12)
plt.colorbar()
plt.xticks(range(len(features)),features,rotation=65,ha='left',fontsize=12)
plt.grid(b=None, which = "major")

plt.tight_layout()
plt.show()

# to output high-resolution image - uncomment the next line
#plt.savefig('PCA_heatmap.png', dpi = 300, bbox_inches = "tight")
```
